### The KID1000+2x3pt data
- https://www.aanda.org/articles/aa/pdf/2021/02/aa39063-20.pdf
- https://kids.strw.leidenuniv.nl/sciencedata.php
- https://kids.strw.leidenuniv.nl/DR4/KiDS-1000_3x2pt_Cosmology.php

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from getdist import plots, MCSamples

kdi1000_2xpt = pd.read_csv('./DES 1Y.csv')
omega_m,sigma_8, weights = kdi1000_2xpt['omega_m'].values, kdi1000_2xpt['sigma_8'].values, kdi1000_2xpt['weights'].values
plt.figure(figsize=(6,6))
plt.scatter(omega_m,sigma_8,s=0.2)
names,labels = ['omega_m','sigma_8'],['omega_m','sigma_8']
# https://getdist.readthedocs.io/en/latest/plot_gallery.html
samples = MCSamples(samples=[omega_m,sigma_8], weights= weights, names = names, labels = labels)
# Triangle plot
plt.figure(figsize=(6,8))
g = plots.get_subplot_plotter()
g.triangle_plot(samples, filled=True)
g = plots.get_single_plotter(width_inch=4, ratio=1)
g.plot_2d(samples, 'omega_m','sigma_8', filled=True)
```

### DES 1Y + 3x2pt

- https://journals.aps.org/prd/abstract/10.1103/PhysRevD.100.023541
- https://ui.adsabs.harvard.edu/abs/2018PhRvD..98d3526A/abstract
- https://arxiv.org/pdf/1708.01530.pdf
- https://desdr-server.ncsa.illinois.edu/despublic/y1a1_files/chains/    data list
- https://des.ncsa.illinois.edu/releases/y1a1

```python
import seaborn as sns
import numpy as np
import pandas as pd
import corner
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", palette="deep")
from astropy.table import Table
from getdist import plots, MCSamples

names,labels = ['omega_m','sigma_8'],['\Omega_m','\sigma_8']
sim_catalog = pd.read_csv('../simulation_paras.csv')
omega_m, sigma_8, weights = sim_catalog['Omega'].values, sim_catalog['Sigm8'].values, np.ones_like(sim_catalog['Sigm8'].values)
samples_sim_catalog= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,)

DES_1Y = pd.read_csv('./DES 1Y.csv')
omega_m, sigma_8, weights = DES_1Y['omega_m'].values, DES_1Y['sigma_8'].values, DES_1Y['weights'].values
# https://getdist.readthedocs.io/en/latest/plot_gallery.html
samples_DES_1Y = MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                           settings={'smooth_scale_2D':0.5})

kid1000 = pd.read_csv('./kid1000.csv')
omega_m, sigma_8, weights = kid1000['omega_m'].values, kid1000['sigma_8'].values, kid1000['weights'].values
samples_kid1000 = MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                            settings={'smooth_scale_2D':0.5})

kid1000_2x3pt = pd.read_csv('./kid1000+2x3pt.csv')
omega_m, sigma_8, weights = kid1000_2x3pt['omega_m'].values, kid1000_2x3pt['sigma_8'].values, kid1000_2x3pt['weights'].values
samples_kid1000_2x3pt = MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})


RF_DR1 = pd.read_csv('../Results/Exclude_2_RF_R1_weight.csv')
omega_m, sigma_8, weights = RF_DR1['Omega'].values, RF_DR1['Sigm8'].values, RF_DR1['weight'].values
samples_RF_DR1= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})

RF_size_DR1 = pd.read_csv('../Results/Exclude_2_RF_size_R1_weight.csv')
omega_m, sigma_8, weights = RF_size_DR1['Omega'].values, RF_size_DR1['Sigm8'].values, RF_size_DR1['weight'].values
samples_RF_size_DR1= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})

RF_eFEDS = pd.read_csv('../Results/Exclude_2_RF_eFEDS_weight.csv')
omega_m, sigma_8, weights = RF_eFEDS['Omega'].values, RF_eFEDS['Sigm8'].values, RF_eFEDS['weight'].values
samples_RF_eFEDS= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})

RF_size_eFEDS = pd.read_csv('../Results/Exclude_2_RF_size_eFEDS_weight.csv')
omega_m, sigma_8, weights = RF_size_eFEDS['Omega'].values, RF_size_eFEDS['Sigm8'].values, RF_size_eFEDS['weight'].values
samples_RF_size_eFEDS= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})

RF_eFEDS_test = pd.read_csv('../Results/RF_test_eFEDS_weight.csv')
omega_m, sigma_8, weights = RF_eFEDS_test['Omega'].values, RF_eFEDS_test['Sigm8'].values, RF_eFEDS_test['weight'].values
samples_RF_eFEDS_test= MCSamples(samples=[omega_m,sigma_8], weights=weights, names=names, labels=labels,
                                  settings={'smooth_scale_2D':0.5})
```

```python
plt.figure(figsize=(16,8),dpi=160)
g = plots.get_subplot_plotter(width_inch=8)
g.plot_2d(  [samples_DES_1Y,samples_kid1000, samples_kid1000_2x3pt, samples_RF_eFEDS, samples_RF_eFEDS_test], 
            'omega_m','sigma_8', 
            filled=True, 
            lims=[0.1, 0.6, 0.5, 1.1])
g.add_legend(['DES_1Y ', 'kid1000', 'kid1000+2x3pt', 'RF_eFEDS', 'RF_eFEDS+size'],colored_text=True, legend_loc='upper right')
g.plot_2d_scatter(samples_sim_catalog, 'omega_m','sigma_8', color='blue', scatter_size=30, lims=[0.1, 0.6, 0.5, 1.1])
```
