```python
import numpy as np
from scipy.special import erf
from astropy.cosmology import Planck18 as cosmo  # 

def L_lim(z, F_lim):
    D_L = cosmo.luminosity_distance(z).cgs.value  # : cm
    return 4 * np.pi * D_L**2 * F_lim

def detection_probability_L(L, z, F_lim=5e-14, sigma_F=1.5e-14):
    D_L = cosmo.luminosity_distance(z).cgs.value
    L_lim_z = 4 * np.pi * D_L**2 * F_lim
    sigma_L = 4 * np.pi * D_L**2 * sigma_F
    return 0.5 * (1 + erf((L - L_lim_z) / (np.sqrt(2) * sigma_L)))

import matplotlib.pyplot as plt

L_vals = np.logspace(41, 45, 200)  # erg/s
zs = [0.05, 0.1, 0.2, 0.4]

for z in zs:
    P_vals = detection_probability_L(L_vals, z)
    plt.plot(L_vals, P_vals, label=f'z={z}')

plt.xscale('log')
plt.xlabel(r'Luminosity $L\ (\mathrm{erg\,s^{-1}})$')
plt.ylabel('Detection Probability')
plt.legend()
plt.title('eROSITA Detection Probability vs. $L$ at Different $z$')
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.show()
```

```python
from MLCI.data_processing import preprocess
import yaml, re
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd
from astropy.table import Table
import numpy as np
from MLCI.utils.config_creator import YamlCreator
from MLCI.visualization.plot import plot_cosmology_paras_vs_test
import matplotlib.pyplot as plt


def calculate_bin_edges(pandas_table, input_column_rangs, bin_num, fix_gird=True):
    bin_edges = {}
    if fix_gird:
        for name, range in input_column_rangs.items():
            bin_edges[name] = np.linspace(range[0], range[1], bin_num+1)
        return bin_edges
    edge_percentages = np.linspace(0, 100, bin_num + 1)
    col_names = list(input_column_rangs.keys())
    for col_name  in col_names:
        data = pandas_table[col_name].values
        # finding the 0%, 10%, 20% ... 90%, 100% points as edges
        bin_edges[col_name] = [np.percentile(data, percentage) for percentage in edge_percentages]
        bin_edges[col_name][-1] = bin_edges[col_name][-1] + 1e-3
    return bin_edges


def obtain_bin(pandas_table, input_column_rangs, bin_edges):
    col_names = list(input_column_rangs.keys())
    bins = [bin_edges[col] for col in col_names]
    values = pandas_table[col_names].values
    hist, edges = np.histogramdd(values, bins=bins)
    hist = hist.flatten().astype(int)
    return hist


def merge_small_bins(obs, sim, min_count=5):

    obs = np.array(obs, dtype=int)
    sim = np.array(sim, dtype=int)
    total = obs + sim
    mask = total >= min_count
    if mask.sum() == len(total):
        return obs, sim
    else:
        obs_new = np.concatenate([obs[mask], [obs[~mask].sum()]])
        sim_new = np.concatenate([sim[mask], [sim[~mask].sum()]])
        return obs_new, sim_new



def calculate_sinificant(model_name, config, opt, bin_num=4, min_num=5, fix_gird=True, weighted=True):
    mulp = 1
    if opt == 'R500-Mgas-L-T-z':
        cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0], 'L':[43, 46], 'T':[-0.2, 1], 'z':[0, 1.0]}
    elif opt == 'R500-Mgas-L-T':
        cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0], 'L':[43, 46], 'T':[-0.2, 1]}
    elif opt == 'R500-Mgas':
        cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0],}
    else:
        raise ValueError('opt not recognized')
    simulated_data = preprocess.simulated_data(fac1=config['noise_level'][0], fac2=config['noise_level'][1])
    simulated_data = preprocess.filtering(simulated_data, cols)
    simulated_data = simulated_data.to_pandas()
    observed_data = preprocess.observed_data()
    observed_data = preprocess.filtering(observed_data, cols)
    observed_data = observed_data.to_pandas()
    bin_edges = calculate_bin_edges(observed_data, cols, bin_num, fix_gird)
    if model_name == 'ref':
        mean_prob = np.ones(15)/15
    elif model_name == 'synthesis':
        mean_prob = np.zeros(15)
        #mean_prob[2] = 0.66
        #mean_prob[9] = 0.33
        mean_prob[13] = 1.0
    elif weighted:
        pred_df = pd.read_csv(f"./results/{model_name}/observation.csv")
        prob_col =  ['prob_' + c for c in config['simulations']]
        prob_val = pred_df[prob_col].values
        confusion_matrix = pd.read_csv(f'figures/{model_name}/CM.csv', index_col=0)
        confusion_matrix = confusion_matrix.values
        new_pro = np.matmul(prob_val, confusion_matrix.T)
        norm_new_pro = new_pro / np.sum(new_pro, axis=1, keepdims=True)
        norm_new_pro = np.nan_to_num(norm_new_pro, nan=0.0)
        mean_prob = np.mean(norm_new_pro, axis=0)
        print('weighted mean prob:', mean_prob)
    else:
        pred_df = pd.read_csv(f"./results/{model_name}/observation.csv")
        prob_col =  ['prob_' + c for c in config['simulations']]
        prob_val = pred_df[prob_col].values
        mean_prob = np.mean(prob_val, axis=0)
    temp_df = []

    for sim_label, prob in zip(config['simulations'], mean_prob):
        subset = simulated_data[simulated_data['label']==sim_label].sample(n=mulp*int(prob*len(observed_data)), replace=True, random_state=42)
        temp_df.append(subset.copy())
    artifitial_df = pd.concat(temp_df, axis=0, ignore_index=True)
    if model_name == 'ref':
        print('Using artifitial uniform distribution')
        for col_name, ranges  in cols.items():
            artifitial_df[col_name] = np.random.uniform(ranges[0], ranges[1], size=len(artifitial_df))
    
    data = [artifitial_df, observed_data]
    labels = ['synthesis', 'observation']
    fname = 'figures/' + re.sub(r'[\\/*?:"<>|]', "_", f"{config['description']}_{model_name}_synthesis_dis.pdf")
    #plot.plot_distribution(data, labels, cols, fname=fname)
    xlabels = ['R_corr', 'Mgas', 'L', 'T']
    fname = 'figures/' + re.sub(r'[\\/*?:"<>|]', "_", f"{config['description']}_{model_name}_synthesis_2ddis.pdf")
    #plot.corner_plot(xlabels, Table.from_pandas(artifitial_df), Table.from_pandas(observed_data), fname=fname)
    obs_bin_data = obtain_bin(observed_data, cols, bin_edges)
    sim_bin_data = obtain_bin(artifitial_df, cols, bin_edges)
    #obs_bin_data, sim_bin_data = merge_small_bins(obs_bin_data, sim_bin_data, min_count=5)
    contingency_table = np.vstack([obs_bin_data, sim_bin_data]) + min_num
    chi2, p, dof, expected = chi2_contingency(contingency_table, correction=False)
    print(model_name, config['description'], 'chi2:', chi2, 'p:', p, 'dof:',dof)
    return chi2, p, dof, expected


def plot_test(fig_name, significant_fname, chi2_opt, ref_chi_name='ref'):
    yaml_creter = YamlCreator(fname='default_test_para.yaml')
    test_config = yaml_creter.load_yaml()
    model_names = ['RFtest1', 'RFtest4', 'RFtest5', 'RFtest31', 'RFtest33', 'RFtest2', 'RFtest24', 'RFtest32', 'RFtest25', #'synthesis',
                'RFtest6', 'RFtest7', 'RFtest19', 'RFtest20', 'RFtest8', 'RFtest9', 'RFtest10', 'RFtest11', 'RFtest12', 'RFtest13',
                'RFtest14', 'RFtest15', 'RFtest17']
    
    synthesis = test_config['RFtest1'].copy()
    synthesis['simulations'] = [f'C{i+1}' for i in range(15)]
    synthesis['description'] = 'synthesis C13'
    test_config['synthesis'] = synthesis.copy()

    show_name = []
    keff = []
    for key in model_names:
        show_name.append(test_config[key]['description'])
        keff.append(len(test_config[key]['simulations']))
    keff = np.zeros_like(keff)
    plot_cosmology_paras_vs_test(model_names, show_name, keff=keff,fname=fig_name, significant=significant_fname, L0=43, chi2_opt=chi2_opt, chi2_range=None, ref_chi_name=ref_chi_name) #[0.8,1.2]
    plt.close()


def get_all_simulation_significant(opt, bin_num, chi2_opt, col_opt):
    file_name = './config/default_test_para.yaml'
    with open(file_name, 'r') as file:
        test_set = yaml.safe_load(file)

    ref = test_set['RFtest1'].copy()
    ref['simulations'] = [f'C{i+1}' for i in range(15)]
    ref['description'] = 'random uniform distribution'
    test_set['ref'] = ref.copy()
    

    synthesis = test_set['RFtest1'].copy()
    synthesis['simulations'] = [f'C{i+1}' for i in range(15)]
    synthesis['description'] = 'synthesis C13'
    test_set['synthesis'] = synthesis.copy()

    sinificants = []
    for model_name, config in test_set.items():
        print('Ready testing:', model_name, config)
        chi2, p, dof, expected = calculate_sinificant(model_name, config, opt=col_opt, bin_num=bin_num, min_num=1, fix_gird=True)
        signifi = {'model_name': model_name, 'des': config['description'], 'chi2':chi2, 'p':p, 'dof':dof}
        sinificants.append(signifi.copy())

    sinificants = Table(sinificants)
    fname = f"results/sinificants_{chi2_opt}_opt={opt}_bin={bin_num}.fits"
    sinificants.write(fname, overwrite=True)
 


bin_list = [5, 10]
opt_list = ['R500-Mgas', 'R500-Mgas-L-T']
chi2_opt_list = ['new_chi2',] #'normal_chi2'
for bin_num in bin_list:    
    for opt in opt_list:
        for chi2_opt in chi2_opt_list:
            print(f"Start testing: fetures={opt}, bin={bin_num}, chi2_opt={chi2_opt}")
            #get_all_simulation_significant(opt, bin_num, chi2_opt, col_opt=opt)
            fig_name =  f"figures/chi2_test/sinificants_{chi2_opt}_opt={opt}_bin={bin_num}.pdf"
            fname = f"results/sinificants_{chi2_opt}_opt={opt}_bin={bin_num}.fits"
            plot_test(fig_name, fname, chi2_opt, ref_chi_name='synthesis')
            #plot_test(fig_name, fname, chi2_opt, ref_chi_name='ref')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def new_plot_distribution(dataset, labels, col_rangs, fname=''):
    colors = ['red', 'green', 'blue']
    vars = col_rangs.keys()
    units = ['$Log_{10}(R/kpc)$', '$Log_{10}(M/M_{sun})$', '$Log_{10}(L/erg \cdot s^{-1})$', '$Log_{10}(T/keV)$', 'z']
    names = ['$R_{500}$', '$M_{gas}$', '$L_{gas}$', '$T_{gas}$', 'z']
    xrangs = col_rangs.values()
    f_size = 14
    fig, axs = plt.subplots(1, 5, figsize=(20, 5), dpi=160)
    for var, name, unit, ax, xrang in zip(vars, names, units, axs, xrangs):
        bw = (xrang[1] - xrang[0])/20
        for label, data, color in zip(labels, dataset, colors):
            label = label
            sns.histplot(data[var], ax=ax, kde=False, label=label, palette='dark', color=color, stat="density", alpha=0.4, binwidth=bw)
            ax.set_yscale('log')
        x0, x1 = np.round(xrang[0],1), np.round(xrang[1],1)
        ax.set_xlabel(unit)
        ax.set_title(f'Density of {name}',fontsize=f_size)
        ax.set_ylabel('')
    axs[4].legend(framealpha=0.1, fontsize=12, loc="upper right") 
    axs[0].set_ylabel('Density', fontsize=f_size)
    plt.xticks(fontsize=f_size)
    plt.yticks(fontsize=f_size)
    if fname == '':
        fname = f'./figures/{fname}_distr.pdf'
    plt.savefig(fname, bbox_inches='tight', pad_inches=0.1)
    plt.close()

new_plot_distribution(data, labels, cols, fname=f"figures/features_dis.pdf")
```

```python
from MLCI.data_processing import preprocess
import yaml, re
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd
from astropy.table import Table
from MLCI.visualization import plot

def calculate_bin_edges(pandas_table, input_column_rangs, bin_num, fix_gird=True):
    bin_edges = {}
    if fix_gird:
        for name, range in input_column_rangs.items():
            bin_edges[name] = np.linspace(range[0], range[1], bin_num+1)
        return bin_edges
    edge_percentages = np.linspace(0, 100, bin_num + 1)
    col_names = list(input_column_rangs.keys())
    for col_name  in col_names:
        data = pandas_table[col_name].values
        # finding the 0%, 10%, 20% ... 90%, 100% points as edges
        bin_edges[col_name] = [np.percentile(data, percentage) for percentage in edge_percentages]
        bin_edges[col_name][-1] = bin_edges[col_name][-1] + 1e-3
    return bin_edges


def obtain_bin(pandas_table, input_column_rangs, bin_edges):
    col_names = list(input_column_rangs.keys())
    bins = [bin_edges[col] for col in col_names]
    values = pandas_table[col_names].values
    hist, edges = np.histogramdd(values, bins=bins)
    hist = hist.flatten().astype(int)
    return hist


def merge_small_bins(obs, sim, min_count=5):

    obs = np.array(obs, dtype=int)
    sim = np.array(sim, dtype=int)

    total = obs + sim
    mask = total >= min_count

    if mask.sum() == len(total):
        #  bin 
        return obs, sim
    else:
        #  bin "" bin
        obs_new = np.concatenate([obs[mask], [obs[~mask].sum()]])
        sim_new = np.concatenate([sim[mask], [sim[~mask].sum()]])
        return obs_new, sim_new
    

def calculate_sinificant(model_name, config, bin_num=4, min_num=5, fix_gird=True, weighted=True):
    mulp = 1
    #cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0], 'L':[43, 46], 'T':[-0.2, 1], 'z':[0, 1.0]}
    #cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0],}
    cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0], 'L':[43, 46], 'T':[-0.2, 1]}
    simulated_data = preprocess.simulated_data(fac1=config['noise_level'][0], fac2=config['noise_level'][1])
    simulated_data = preprocess.filtering(simulated_data, cols)
    simulated_data = simulated_data.to_pandas()
    observed_data = preprocess.observed_data()
    observed_data = preprocess.filtering(observed_data, cols)
    observed_data = observed_data.to_pandas()
    bin_edges = calculate_bin_edges(observed_data, cols, bin_num, fix_gird)
    if model_name == 'ref':
        mean_prob = np.ones(15)/15
    elif model_name == 'C3':
        mean_prob = np.zeros(15)
        mean_prob[2] = 0.66
        mean_prob[9] = 0.33
    elif weighted:
        pred_df = pd.read_csv(f"./results/{model_name}/observation.csv")
        prob_col =  ['prob_' + c for c in config['simulations']]
        prob_val = pred_df[prob_col].values
        confusion_matrix = pd.read_csv(f'figures/{model_name}/CM.csv', index_col=0)
        confusion_matrix = confusion_matrix.values
        new_pro = np.matmul(prob_val, confusion_matrix.T)
        norm_new_pro = new_pro / np.sum(new_pro, axis=1, keepdims=True)
        norm_new_pro = np.nan_to_num(norm_new_pro, nan=0.0)
        mean_prob = np.mean(norm_new_pro, axis=0)
        print('weighted mean prob:', mean_prob)
    else:
        pred_df = pd.read_csv(f"./results/{model_name}/observation.csv")
        prob_col =  ['prob_' + c for c in config['simulations']]
        prob_val = pred_df[prob_col].values
        mean_prob = np.mean(prob_val, axis=0)
    temp_df = []
    for sim_label, prob in zip(config['simulations'], mean_prob):
        subset = simulated_data[simulated_data['label']==sim_label].sample(n=mulp*int(prob*len(observed_data)), replace=True, random_state=42)
        temp_df.append(subset.copy())
    artifitial_df = pd.concat(temp_df, axis=0, ignore_index=True)
    data = [artifitial_df, observed_data]
    labels = ['synthesis', 'observation']
    fname = 'figures/' + re.sub(r'[\\/*?:"<>|]', "_", f"{config['description']}_{model_name}_synthesis_dis.pdf")
    #plot.plot_distribution(data, labels, cols, fname=fname)
    xlabels = ['R_corr', 'Mgas', 'L', 'T']
    fname = 'figures/' + re.sub(r'[\\/*?:"<>|]', "_", f"{config['description']}_{model_name}_synthesis_2ddis.pdf")
    #plot.corner_plot(xlabels, Table.from_pandas(artifitial_df), Table.from_pandas(observed_data), fname=fname)
    obs_bin_data = obtain_bin(observed_data, cols, bin_edges)
    sim_bin_data = obtain_bin(artifitial_df, cols, bin_edges)
    #obs_bin_data, sim_bin_data = merge_small_bins(obs_bin_data, sim_bin_data, min_count=5)
    contingency_table = np.vstack([obs_bin_data, sim_bin_data]) + min_num
    chi2, p, dof, expected = chi2_contingency(contingency_table, correction=False)
    print(model_name, config['description'], 'chi2:', chi2, 'p:', p, 'dof:',dof)
    return chi2, p, dof, expected


file_name = './config/default_test_para.yaml'
with open(file_name, 'r') as file:
    test_set = yaml.safe_load(file)

ref = test_set['RFtest1'].copy()
ref['simulations'] = [f'C{i+1}' for i in range(15)]
test_set['ref'] = ref.copy()

C3 = test_set['RFtest1'].copy()
C3['simulations'] = [f'C{i+1}' for i in range(15)]
test_set['C3'] = C3.copy()

sinificants = []
bin_num = 10
for model_name, config in test_set.items():
    #if model_name == 'RFtest31' or model_name == 'RFtest33':
        
    print('Ready testing:', model_name, config)
    #if model_name == "RFtest21":
    chi2, p, dof, expected = calculate_sinificant(model_name, config, bin_num=bin_num, min_num=1, fix_gird=True)
    signifi = {'model_name': model_name, 'des': config['description'], 'chi2':chi2, 'p':p, 'dof':dof}
    sinificants.append(signifi.copy())


sinificants = Table(sinificants)
sinificants.write(f"results/sinificants_no_z_bin{bin_num}.fits", overwrite=True)
```
