```python
import numpy as np
from scipy.special import erf
from astropy.cosmology import Planck18 as cosmo  # 

def L_lim(z, F_lim):
    D_L = cosmo.luminosity_distance(z).cgs.value  # : cm
    return 4 * np.pi * D_L**2 * F_lim

def detection_probability_L(L, z, F_lim=5e-14, sigma_F=1.5e-14):
    D_L = cosmo.luminosity_distance(z).cgs.value
    L_lim_z = 4 * np.pi * D_L**2 * F_lim
    sigma_L = 4 * np.pi * D_L**2 * sigma_F
    return 0.5 * (1 + erf((L - L_lim_z) / (np.sqrt(2) * sigma_L)))

import matplotlib.pyplot as plt

L_vals = np.logspace(41, 45, 200)  # erg/s
zs = [0.05, 0.1, 0.2, 0.4]

for z in zs:
    P_vals = detection_probability_L(L_vals, z)
    plt.plot(L_vals, P_vals, label=f'z={z}')

plt.xscale('log')
plt.xlabel(r'Luminosity $L\ (\mathrm{erg\,s^{-1}})$')
plt.ylabel('Detection Probability')
plt.legend()
plt.title('eROSITA Detection Probability vs. $L$ at Different $z$')
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.show()
```

```python
from MLCI.visualization import comparision, plot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
from astropy.cosmology import Planck18


delta_cosmology = []
z_0, z_1 = 0.1, 0.8
model_names = ['RFtest13', 'RFtest20','RFtest7', 'RFtest1', 'RFtest6', 'RFtest19', 'RFtest12']
for model_name in model_names:
    results = pd.read_csv(f"./results/{model_name}/observation.csv")
    subset = results[(results['z'] >= z_0) & (results['z'] < z_1)]
    cosmology_para = plot.get_expectation_1sigam(subset, model_name, model_name)
    delta_cosmology.append(cosmology_para)

ref_cosmology = delta_cosmology[2]
#ref_cosmology = {'Omega_medians':Planck18.Om0, 'Sigm8_medians':Planck18.meta['sigma8'], 'Hubble_medians':Planck18.H0.value/100, 'OmegaB_medians':Planck18.Ob0}
delta_cosmology = Table(delta_cosmology)

def plot_cosmology_paras_vs_delta(delta_cosmology, ref_cosmology, fname='test'):
    para_names = ['Omega', 'Sigm8', 'Hubble', 'OmegaB']
    label_names = ['$\Omega_m$', '$\sigma_8$', '$h_0$', '$\Omega_B$']
    plt.figure(figsize=(8, 5))
    xoffsets = np.linspace(-0.1, 0.1, len(para_names))
    for para_name, label, x_off in zip(para_names, label_names, xoffsets):
        x = np.arange(len(delta_cosmology)) + x_off
        y = delta_cosmology[f'{para_name}_medians']
        yerr = np.array([delta_cosmology[f'{para_name}_lower_errors'], delta_cosmology[f'{para_name}_upper_errors']])
        norm_value = ref_cosmology[f'{para_name}_medians']
        plt.errorbar(x, y/norm_value, yerr=yerr/norm_value, fmt='o', capsize=4, label=label)
    show_name = ['-0.3', '$RF_{-2 \delta}$', '$RF_{-\delta}$', 'RF', '$RF_{+\delta}$', '$RF_{+2 \delta}$', '+0.3']
    plt.xticks(ticks=np.arange(len(delta_cosmology)), labels=show_name)
    #plt.xticks(ticks=np.arange(len(delta_cosmology)), labels=delta_cosmology['Model Name'].data)
    plt.ylabel('shiftded / No shifted')
    plt.legend()
    plt.axhline(y=1, color='red', linestyle='--', linewidth=1)
    #plt.title('Cosmological Parameters vs feedback, reference: planck2018')
    plt.title('Cosmological Parameters vs feedback')
    plt.legend(title='Parameters')
    plt.tight_layout()
    plt.savefig(f'./figures/{fname}_Cosmological_Parameters_vs_delta_new.pdf')

    '''
    for ky in para_names:
        for row in delta_cosmology:
            median_shift = row[f'{ky}_medians'] - ref_cosmology[f'{ky}_medians']
            lower_errors_shift = row[f'{ky}_lower_errors']-ref_cosmology[f'{ky}_lower_errors']
            upper_errors_shift = row[f'{ky}_upper_errors']-ref_cosmology[f'{ky}_upper_errors']
            print(f"{row['Model Name']}: Delta {ky} = {median_shift:.4f}")
            print(f"Delta(-sigma) = {lower_errors_shift:.3f}, Delta(+sigma) = {upper_errors_shift:.3f} ")
    '''

plot_cosmology_paras_vs_delta(delta_cosmology, ref_cosmology, fname='test')
```

```python
from astropy.table import Table
from astropy.coordinates import SkyCoord
import astropy.units as u

results = []
keys = ['name1', 'name2', 'ra_h', 'ra_m', 'ra_s', 'dec_d', 'dec_m', 'dec_s', 'cz', 'Dispersione', 'Velocita Della BCG', 'Differenza v(bcg)-redshift/sigma']
with open('./Data/new_vdis_catalog/galcen.txt', 'r', encoding='utf-8') as f:
    for line in f:
        #print(line.strip())
        values = line.strip().split()
        data_dict = dict(zip(keys, values))
        results.append(data_dict)

results = Table(results)

results['ra_s'] = [float(str(s).strip().rstrip('.')) for s in results['ra_s']]
results['dec_s'] = [float(str(s).strip().rstrip('.')) for s in results['dec_s']]

#results['ra'] = 15 * (results['ra_h'].data.astype(float) + results['ra_m'].data.astype(float)/60 + results['ra_s'].data.astype(float)/3600)
#results['dec'] = results['dec_d'].data.astype(float) +  results['dec_m'].data.astype(float)/60 + results['dec_s'].data.astype(float)/3600

ra_h, ra_m, ra_s = results['ra_h'].data.astype(float), results['ra_m'].data.astype(float), results['ra_s'].data.astype(float)
dec_d, dec_m, dec_s = results['dec_d'].data.astype(float), results['dec_m'].data.astype(float), results['dec_s'].data.astype(float)


coord = SkyCoord(ra=ra_h*u.hour + ra_m*u.minute + ra_s*u.second,
                 dec=dec_d*u.deg + dec_m*u.arcmin + dec_s*u.arcsec)

results['ra'] = coord.ra.deg
results['dec'] = coord.dec.deg
results['name'] = results['name1'] + results['name2']
results['Dispersione'] = results['Dispersione'].data.astype(float)

results.write(f'./Data/new_vdis_catalog/new_dispersion_catalog.fits', format="fits", overwrite=True)
```

```python
#import preprocess, boosting, comparision, model, plot, features_importance, agn_feedback
from MLCI.data_processing import preprocess, boosting
from MLCI.visualization import plot
import numpy as np
import yaml

file_name = './config/default_test_para.yaml'
with open(file_name, 'r') as file:
    test_set = yaml.safe_load(file)
config = test_set['RFtest1']
cols = {config['R_type']:[2.5, 3.5], 'Mgas':[12.0, 14.0], 'L':[43, 46], 'T':[-0.2, 1], 'z':[0, 1.0]}
vars_name = list(cols.keys())
simulated_data=preprocess.simulated_data(fac1=config['noise_level'][0], fac2=config['noise_level'][1])
exclude_set = ['HR', 'C8a1', 'C8a2']
mask = ~np.isin(simulated_data['label'], exclude_set)
simulated_data = simulated_data[mask]
observed_data = preprocess.observed_data()
observed_data = preprocess.filtering(observed_data, cols)
simulated_data= preprocess.filtering(simulated_data, cols)
# boosting data
observed_boost_info, observed_boost_sample = boosting.boosting(observed_data, vars_name, num=config['sample_num'], sampler='multinorm')
simulated_boost_info, simulated_boost_sample = boosting.boosting(simulated_data, vars_name, num=config['sample_num'], sampler='multinorm', max_num=config['max_num'])
# show after boosting distribute
observed_boost_sample = preprocess.filtering(observed_boost_sample, cols)
simulated_boost_sample= preprocess.filtering(simulated_boost_sample, cols)
eFEDS = observed_boost_sample[observed_boost_sample['label'] == 'eFEDS']
eRASS1 = observed_boost_sample[observed_boost_sample['label'] == 'eRASS1']
data = [eFEDS, eRASS1, simulated_boost_sample]
labels = ['eFEDS', 'eRASS1', 'Simualtions']
output_line = f"Dataset_num: {labels} {len(eFEDS)}-{len(eRASS1)}-{len(simulated_boost_sample)}"
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def new_plot_distribution(dataset, labels, col_rangs, fname=''):
    colors = ['red', 'green', 'blue']
    vars = col_rangs.keys()
    units = ['$Log_{10}(R/kpc)$', '$Log_{10}(M/M_{sun})$', '$Log_{10}(L/erg \cdot s^{-1})$', '$Log_{10}(T/keV)$', 'z']
    names = ['$R_{500}$', '$M_{gas}$', '$L_{gas}$', '$T_{gas}$', 'z']
    xrangs = col_rangs.values()
    f_size = 14
    fig, axs = plt.subplots(1, 5, figsize=(20, 5), dpi=160)
    for var, name, unit, ax, xrang in zip(vars, names, units, axs, xrangs):
        bw = (xrang[1] - xrang[0])/20
        for label, data, color in zip(labels, dataset, colors):
            label = label
            sns.histplot(data[var], ax=ax, kde=False, label=label, palette='dark', color=color, stat="density", alpha=0.4, binwidth=bw)
            ax.set_yscale('log')
        x0, x1 = np.round(xrang[0],1), np.round(xrang[1],1)
        ax.set_xlabel(unit)
        ax.set_title(f'Density of {name}',fontsize=f_size)
        ax.set_ylabel('')
    axs[4].legend(framealpha=0.1, fontsize=12, loc="upper right") 
    axs[0].set_ylabel('Density', fontsize=f_size)
    plt.xticks(fontsize=f_size)
    plt.yticks(fontsize=f_size)
    if fname == '':
        fname = f'./figures/{fname}_distr.pdf'
    plt.savefig(fname, bbox_inches='tight', pad_inches=0.1)
    plt.close()

new_plot_distribution(data, labels, cols, fname=f"figures/features_dis.pdf")
```
